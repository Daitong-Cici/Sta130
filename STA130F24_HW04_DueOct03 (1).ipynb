````python
import numpy as np

# Function to compute a bootstrap confidence interval for the population mean
def bootstrap_ci_mean(sample, num_bootstrap=10000, confidence_level=0.95):
    #replace the np.mean(bootstrap_sample) line with np.median(bootstrap_sample)
    """
    Compute a bootstrap confidence interval for the population mean.
    
    Parameters:
    sample (array-like): Sample data.
    num_bootstrap (int): Number of bootstrap samples to generate (default=10000).
    confidence_level (float): Confidence level (default=0.95).
    
    Returns:
    tuple: Lower and upper bounds of the confidence interval.
    """
    # Step 1: Create an array to hold the bootstrap sample means
    bootstrap_means = np.empty(num_bootstrap)
    # we could replace means with medians
    
    # Step 2: Resample the original sample with replacement
    for i in range(num_bootstrap):
        bootstrap_sample = np.random.choice(sample, size=len(sample), replace=True)
        bootstrap_means[i] = np.mean(bootstrap_sample)  # Mean of bootstrap sample
    
    # Step 3: Calculate the lower and upper percentiles for the CI
    lower_percentile = (1 - confidence_level) / 2 * 100
    upper_percentile = (1 + confidence_level) / 2 * 100
    
    lower_bound = np.percentile(bootstrap_means, lower_percentile)
    upper_bound = np.percentile(bootstrap_means, upper_percentile)
    
    return lower_bound, upper_bound
# we could replace means with median

# Example: Generating a sample and calculating the 95% CI for the mean
sample = np.array([1.2, 2.5, 3.3, 4.1, 5.6, 6.8, 7.9, 8.3, 9.1])
ci_mean = bootstrap_ci_mean(sample)
print(f"95% Bootstrap Confidence Interval for the Mean: {ci_mean}")
#we could replace mean with median
````

If we want to change the code to produce a 95% bootstrap confidence interval for different population parameter (other than the population mean, such as the population median), we just need to replace the np.mean(bootstrap_sample) line with np.median(bootstrap_sample). The population parameters are based on all of the original dataset but the sample statistic is based on the sample that we picked from the dataset.When it comes to the confidence intervals, we can estimate the population parameter with the sample statistic because it is the central point around which the confidence interval constructed. We rarely have access to the entire population, so we use sample data to estimate the population parameter. Since different samples will produce different sample statistics (e.g., different sample means), the confidence interval accounts for the uncertainty in the estimation process. The sample statistic is just one possible estimate from many potential samples. The sample statistic varies from sample to sample due to random sampling variability. This variability is reflected in the confidence interval, which recognizes that our estimate (the sample statistic) could differ from the true population parameter. The confidence interval doesnt describe the probability of the population parameter being within the interval (because the population parameter is fixed). Instead, it describes the confidence we have that, if we were to take many samples, 95% of the confidence intervals constructed from those samples would contain the true population parameter.

````python
import numpy as np

# Function to perform bootstrapping
def bootstrap_mean_diff(initial_scores, final_scores, n_iterations=10000):
    mean_diffs = []
    for _ in range(n_iterations):
        boot_initial = np.random.choice(initial_scores, size=len(initial_scores), replace=True)
        boot_final = np.random.choice(final_scores, size=len(final_scores), replace=True)
        mean_diff = np.mean(boot_final) - np.mean(boot_initial)
        mean_diffs.append(mean_diff)
    return np.array(mean_diffs)

# Perform bootstrapping
initial_scores = df['InitialHealthScore'].values
final_scores = df['FinalHealthScore'].values
bootstrap_diffs = bootstrap_mean_diff(initial_scores, final_scores)

# Calculate the confidence interval
lower_ci = np.percentile(bootstrap_diffs, 2.5)
upper_ci = np.percentile(bootstrap_diffs, 97.5)

# Calculate the observed mean difference
observed_mean_diff = np.mean(final_scores) - np.mean(initial_scores)

print(f'Observed Mean Difference: {observed_mean_diff}')
print(f'95% Confidence Interval: [{lower_ci}, {upper_ci}]')
````
