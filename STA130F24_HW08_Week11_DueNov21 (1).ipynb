### **Summary**

#### (a) **What is a Classification Decision Tree?**
- A **Classification Decision Tree** predicts **categorical outcomes** (e.g., Yes/No, Spam/Not Spam) by splitting data into groups based on feature conditions.
- **Examples**:
  - Medical Diagnosis (e.g., predicting cancer type).
  - Fraud Detection (e.g., classifying a transaction as fraud or legitimate).
  - Credit Risk Assessment (e.g., classifying applicants as low-risk or high-risk).
- **Difference from Regression**: Classification predicts categories, while Regression predicts continuous numerical values.

#### (b) **How Does It Work?**
- **Sequential Decisions**: The tree asks feature-based questions (e.g., "Is Age > 30?"), splitting data at each node.
- **Final Prediction**: At the leaf node, the majority class is assigned as the prediction.

#### **Comparison to Multiple Linear Regression**:
- Decision Trees use **conditional splits** to predict categories.
- Linear Regression uses a **mathematical equation** to predict a continuous value.
- Decision Trees handle **nonlinear relationships** and interactions naturally, unlike Linear Regression, which assumes linearity.

1. **Accuracy** measures the proportion of true results (both true positives and true negatives) in the population.

$$\text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN}$$
    
2. **Sensitivity** measures the proportion of actual positives that are correctly identified.

$$\text{Sensitivity} = \frac{TP}{TP + FN}$$

3. **Specificity** measures the proportion of actual negatives that are correctly identified.

$$\text{Specificity} = \frac{TN}{TN + FP}$$

4. **Precision** measures the proportion of positive identifications that were actually correct.

$$\text{Precision} = \frac{TP}{TP + FP}$$

### **Summary**

You don’t need to memorize the formulas, but understanding the **concepts** behind TP, TN, FP, and FN is crucial. Here's a quick breakdown:

1. **True Positive (TP)**: Correctly identified positive cases.  
   - Example: The model says "disease" when the patient actually has the disease.

2. **True Negative (TN)**: Correctly identified negative cases.  
   - Example: The model says "no disease" when the patient is healthy.

3. **False Positive (FP)**: Incorrectly predicted as positive.  
   - Example: The model says "disease" when the patient is actually healthy (Type I Error).

4. **False Negative (FN)**: Incorrectly predicted as negative.  
   - Example: The model says "no disease" when the patient actually has the disease (Type II Error).

---

### **When to Memorize Formulas**
- For exams or interviews: Memorize these key metrics:
  - **Accuracy**: \((TP + TN) / (TP + TN + FP + FN)\)
  - **Sensitivity**: \(TP / (TP + FN)\)
  - **Specificity**: \(TN / (TN + FP)\)
  - **Precision**: \(TP / (TP + FP)\)

````python
# Import necessary libraries
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, recall_score
import pandas as pd
import numpy as np

 
ab_reduced_noNaN = pd.DataFrame(data)

# Split the data into training and testing sets
ab_reduced_noNaN_train, ab_reduced_noNaN_test = train_test_split(
    ab_reduced_noNaN, test_size=0.2, random_state=42
)

# Train clf (List Price only)
X_train_clf = ab_reduced_noNaN_train[['List Price']]
y_train_clf = pd.get_dummies(ab_reduced_noNaN_train["Hard_or_Paper"])['Hard']

clf = DecisionTreeClassifier(max_depth=2, random_state=42)
clf.fit(X_train_clf, y_train_clf)

# Train clf2 (NumPages, Thick, List Price)
X_train_clf2 = ab_reduced_noNaN_train[['NumPages', 'Thick', 'List Price']]
y_train_clf2 = y_train_clf

clf2 = DecisionTreeClassifier(max_depth=4, random_state=42)
clf2.fit(X_train_clf2, y_train_clf2)

# Predictions for clf and clf2
X_test_clf = ab_reduced_noNaN_test[['List Price']]
X_test_clf2 = ab_reduced_noNaN_test[['NumPages', 'Thick', 'List Price']]
y_test = pd.get_dummies(ab_reduced_noNaN_test["Hard_or_Paper"])['Hard']

y_pred_clf = clf.predict(X_test_clf)
y_pred_clf2 = clf2.predict(X_test_clf2)

# Confusion matrices
conf_matrix_clf = confusion_matrix(y_test, y_pred_clf)
conf_matrix_clf2 = confusion_matrix(y_test, y_pred_clf2)

# Metrics calculation
def calculate_metrics(conf_matrix, y_true, y_pred):
    TN, FP, FN, TP = conf_matrix.ravel()
    sensitivity = recall_score(y_true, y_pred)  # TP / (TP + FN)
    specificity = TN / (TN + FP) if (TN + FP) > 0 else 0.0  # TN / (TN + FP)
    accuracy = accuracy_score(y_true, y_pred)  # (TP + TN) / Total
    return sensitivity, specificity, accuracy

# Metrics for clf
sensitivity_clf, specificity_clf, accuracy_clf = calculate_metrics(conf_matrix_clf, y_test, y_pred_clf)
# Metrics for clf2
sensitivity_clf2, specificity_clf2, accuracy_clf2 = calculate_metrics(conf_matrix_clf2, y_test, y_pred_clf2)

# Display results as DataFrame
results = {
    "Model": ["clf", "clf2"],
    "Sensitivity": [np.round(sensitivity_clf, 3), np.round(sensitivity_clf2, 3)],
    "Specificity": [np.round(specificity_clf, 3), np.round(specificity_clf2, 3)],
    "Accuracy": [np.round(accuracy_clf, 3), np.round(accuracy_clf2, 3)],
}

results_df = pd.DataFrame(results)

# Print the metrics results
print("Confusion Matrix Metrics Results:")
print(results_df)
````

<details class="details-example"><summary style="color:blue"><u>Further Guidance</u></summary>
  
> The way a **classification decision tree** is fit is that at each step in the construction process of adding a new **decision node splitting rule** to the current **tree structure**, all possible **decision rules** for all possible **predictor variables** are considered, and the combination that improves the **prediction** the most (as measured by the criterion of either "Gini impurity" or "Shannon entropy") and in accordance with the rules of the decision tree (such as the `max_depth` argument) is added to the **classification decision tree**.  Thus overall "criterion" noted above improves with each new **decision node splitting rule**, so the improvement can thus be tracked and the improvement contributions attributed to the **feature** upon which the **decision node splitting rule** is based.  This means the relative contribution of each **predictor variable** to the overall explanatory power of the model can be calculated, and this is what the `.feature_importances_` attribute does. 
>
> Compared to the simplicity of understanding how different **covariates** contribute towards the final **predicted values** of **multiple linear regression models** (by just reading off the equation to see how predictions work), the the complexity of how all the different **features** interact and combine to together to create the final **predictions** from **classification decision trees** can be staggering. But the so-called **feature importance** heuristics allows us to judge how relatively important the overall contributions from different features are in the final decision tree predictions. Now we just need to be sure we're not **overfitting** our **classification decision trees** since they can be so **complex**. Fortunately, the "GridSearchCV" methodology mentioned in regards to finding the best `max_depth` setting for a tree is going to provide a general answer to the challenge of complexity and **overfitting** in **machine learning models** that is not too hard to understand (and which you might already have some guesses or a hunch about).
